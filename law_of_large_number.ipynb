## Law of large numbers

### 1) For 10 flips

```python
#for 10 flips
import matplotlib.pyplot as plt
import random 

n=1
prob=[]
flip=[]

while n<10:
    head=0
    tail=0

    for i in range(n):
        if random.randint(0,1)==0:
            head+=1
        else:
            tail+=1
    k=head/(head+tail)
    prob.append(k)
    flip.append(n)
    n+=1
```

```python
plt.figure(figsize = (10, 15))
plt.subplot(2,1,1)
plt.hist(prob,100,label='Probality of Heads')
plt.legend()
```

```python
plt.figure(figsize = (10, 15))
plt.subplot(2,1,2)
plt.plot(flip,prob)
plt.xlabel('Number of Flips')
plt.ylabel('Probability of Heads')
plt.grid(True)
plt.show
```
